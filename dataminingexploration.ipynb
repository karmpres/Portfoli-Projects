**Performing year-over-year analysis of expenses for the following interesting independent variables: `Agency Name`, `Summary Object Name`, `Sub Object Name`, and `Vendor Names`. Finding if there is average increase/decrease? Also Identifying outliers.** 

```python
# This shows the count of how many times different vendor have traded. 
fy2019['Vendor Name'].value_counts()
```

```python
# Visual representatation of sum of all the vendor name, won't be pretty.
fy2019.groupby('Vendor Name')['Amount'].agg(np.sum).hist(bins=50)
```

```python
# Looking more closely at the sum, mean, max, min amount of transactiontion
# of  each vendor makes.
vendor_name19 = fy2019.groupby('Vendor Name').Amount.agg(['sum','mean','median','std','max','min'])
vendor_name19.head(10)
```

```python
vendor_name18 = fy2018.groupby('Vendor Name').Amount.agg(['sum','mean','median','std','max','min'])
vendor_name18.head(10)
```

```python
# lets ge tthe difference of amount of 2019 and 2018 with respect to the 
# vendor name
vendor_join = vendor_name18.merge(vendor_name19, on='Vendor Name', how = 'outer',suffixes=('18','19')).fillna(0)
diff_amount_vendor= vendor_join['sum18']-vendor_join['sum19']
vendor_join['diff_amount_vendor']=diff_amount_vendor
vendor_join[['sum18','sum19','diff_amount_vendor']].head(15)
```

```python
fy2018.groupby('Fund Name').Amount.sum().plot.bar(colors='green',figsize=(15,15),position=1,label='2018')
fy2019.groupby('Fund Name').Amount.sum().plot.bar(colors='red',figsize=(15,15),position=0,label='2019')
plt.xlabel('Fund Name')
plt.ylabel('Amount')
plt.legend()
```

```python
print(vendor_name18.max())
print(vendor_name18.min())
print(vendor_name18.sum())
```

```python
# printing out the maximum sum, mean.. values for all the vendors.
print(vendor_name19.max())
print(vendor_name19.min())
print(vendor_name19.sum())
```

** Result Finding for Vendor Name**

1) We can see the diiference of amount between 2019 and 2018, and comparing the total sum of amount of year 2018 and 2019, 9.821682e+09 and 1.030868e+10 respectively. we can see that, the amount of expenses has increased in 2019.

2) Also, we can look at the total sum of standardeviation between 2018 and 2019, 2.981545e+08 and 3.218245e+08 respectively. This also suggests that the data is more spread out in 2019 year.

3) It looks like the data is skewed to right and left. But, it seeems the data is skewd to the right. ALthough, it is very unclear as there are lot of rows from the plot. Since, mean> median we are going to assume it is skewed to the right

4) We can that  the highest amount expenses was from Vendor name of Awards, amount 1.462798e+09 for 2018 and 1.443667e+09 for 2019.

5) We can that  the lowest amount expenses was from Vendor name, amount -4.464460e+07 for 2018 and -9.621943e+06 for 2019. Note that this min is sum of vendor name all expenses.

6) Also note that these values(4)(5) lies above outlier zone.
